### Código de inciante - Machine Learning - ExtraTreesClassifier

```python
import pandas as pd

arquivo = pd.read_csv("winequalityN.csv")
arquivo.head()
```

```python
arquivo = arquivo.dropna()
```

```python
arquivo.tail()
```

```python
print(f"quantidade de reds {arquivo['type'].str.count('red').sum()}\nquantidade de whites {arquivo['type'].str.count('white').sum()}\nquantidade total: {len(arquivo)} vs total_counts: {arquivo['type'].str.count('red').sum() + arquivo['type'].str.count('white').sum()}")
```

```python
arquivo['type'] = arquivo['type'].replace('red', 0)
arquivo['type'] = arquivo['type'].replace('white', 1)
```

```python
#Separando variáveis preditoras e alvo:

y = arquivo['type']
x = arquivo.drop('type', axis = 1)
y
```

```python
from sklearn.model_selection import train_test_split

#Criando conjunto de dados de treino e teste:
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,test_size = 0.3) #30% dos dados servirá para teste
```

```python
from sklearn.ensemble import ExtraTreesClassifier

#Criação do modelo
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)
```

```python
resultado = modelo.score(x_teste, y_teste)
print(f"Acurácia: {resultado}")
```

```python
previsoes = modelo.predict(x_teste)
```
